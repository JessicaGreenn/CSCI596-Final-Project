```python
import matplotlib.pyplot as plt
import numpy as np

n_threads = np.array([1, 2, 4, 8, 16])
small_dataset = np.array(
    [2.4494471549987793, 1.2521562576293945, 1.0895359516143799, 1.184403896331787, 1.5739750862121582])
medium_dataset = np.array(
    [36.56010103225708, 5.2411110401153564, 3.745589017868042, 2.986539125442505, 3.4592387676239014])
large_dataset = np.array(
    [65.6125340461731, 27.224717140197754, 7.222593069076538, 7.175553798675537, 7.586134910583496])

plt.plot(n_threads, small_dataset, label='small dataset', color='b', linestyle='-')
plt.plot(n_threads, medium_dataset, label='medium dataset', color='r', linestyle='--')
plt.plot(n_threads, large_dataset, label='large dataset', color='g', linestyle='-.')
# add title and x-y axis
plt.xlabel('number of threads')
plt.ylabel('time')
plt.legend()
plt.show()
```
